    data.info() showed us a few features that we categorized as the wrong datatypes. 

```python
#convert touch points to int datatype
df[['Native', 'Social', 'TravelEndemic','DSP', 'DSPTravel']] = df[['Native', 'Social', 'TravelEndemic','DSP', 'DSPTravel']].fillna(0).astype(int)
```

```python
#drop columns we don't need
df = df.drop(['Event Timestamp', 'DSP', 'Impression / Click Time'], axis = 1)
```

    In the past we've had problem with bots clicking on ads thousands of times. Are there any bots in our data?

```python
#How many unique people are in our dataset
df['Unique ID'].nunique()
```

```python
#Check if there is any one person with 'too many' touches
publisher_type = df[['Unique ID','Native', 'Social', 'TravelEndemic', 'DSPTravel']].groupby('Unique ID').sum()

#four each column other than Unique ID in the publisher_type, check the max touches
publisher_type = publisher_type.sort_values(['TravelEndemic'],ascending = False)

publisher_type.head(5)
```

```python
#or here's a faster way
publisher_type.max()
```
